# `Types of Data Structures:`
* `List` | `Tuple`
* `Set` | `Dictionary`

|Data Structures type|Mutable|Comments|Indexing|Ordered|Duplicacy|
|----|----|----|----|----|----|
tuple ()|immutable|immutable version of list| possible|yes |allowed |
list []|mutable|-|possible |yes | allowed |
set {}|mutable| -|not |no |not |
dict {key:value}|mutable|-|possible |no |not  |

üí´ `immutable` => `can't be changed` <br>
üí´ `mutable` => `can be changed`

```python
# List (Mutable)
my_list = [1, 2, 3]
my_list[0] = 10

# Tuple (Immutable)
my_tuple = (1, 2, 3)
# my_tuple[0] = 10  # error varum

# Set (Mutable, no duplicate)
my_set = {1, 2, 3}
my_set.add(4)

# Dictionary (Mutable, key:value)
my_dict = {"name": "Ali", "age": 20}
my_dict["age"] = 21

print(my_list)
print(my_tuple)
print(my_set)
print(my_dict)
```

### Create colors = [‚Äòred‚Äô,‚Äôblue‚Äô,‚Äôorange‚Äô], and fruits = [‚Äòorange‚Äô,‚Äôgrapes‚Äô,‚Äôapples‚Äô]

```python
colors = ['red','blue','orange']
fruits = ['orange','grapes','apples']
```

Print color and fruits

```python
print(colors)
print(fruits)
```

Create colors_set, and fruits_set. (using set() ) and print them

```python
colors_set = set(colors)
fruits_set = set(fruits)
print(colors_set)
print(fruits_set)
```

Find the union of both the sets.

```python
union_set = colors_set.union(fruits_set)
print(union_set)
```

Find the intersection of both the sets

```python
intersection_set = colors_set.intersection(fruits_set)
print(intersection_set)
```

Find the elements which are Fruits but not colors (using set.difference() )

```python
fruits_only = fruits_set.difference(colors_set)
print(fruits_only)
```
